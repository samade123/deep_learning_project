# Deep Learning basic explanantion

In a deep learning model it can be broken into at minimum 3 layers of data representation. Layers are made of nodes, which hold some sort of value(data representation)

- The input layer -  this consists fthe same number of nodes as data vales in the input data, so for image classifiacation, the input layer would have the same number node as pixels in the image, 1 node for each pixel. Befeore data enters this layer. It would undergo some sort of Data preprocessing. This would include some sort of transformation  of the data in order to make it easier for the model to "learn" from the data. For Deep learning models due to their greaater number of layers, not that much preprocessign is required. In single layer machine learning models we would likely undergo some feature engineering on nour data, which is using specialist knowledge to transform the data before entering the nearal network to make the model more succesful. Modern day neural networks require less feature engineering as they are more accurate at extracting features from the data presented.

- The output layer - This holds as many nodes as output labels required. In our fashion mnist example, we have 10 output nodes 1 for each possible category(class) label a photo could possibly be. In a classification model, each node will hold a probalbilty value between 0 and 1 of the lilihood of that the iamge will be of that class. The sum of all nodes will equal to 1.

- Hidden Layers - The layers inbetween the input and output layers can be consisdered "hidden" layers. The number of nodes in these layers are deceide by the machne leaerning engineer, to achieve the most effective result. In the example above the layers can be considered "dense" which means all the nodes from the previous layer are connected to all the nodes in the next layer.In the forward feed stage of the model,  the value of the node is determeined by the sum of all the nodes and  weights associated at that node. a bias value specific to the layer, is also added and and the whole sum is is wrapped inside and activation function. the type of function is decieded nby the engineer and is based on what type of value the user would like to see at the output neuron.

# A more in depth explanantion of neurons


> $Y =(\sum_{i=0}^n w_i x_i) + b$

> $w$ stands for the weight of each connection to the neuron

> $x$ stands for the value of the connected neuron from the previous value

> $b$ stands for the bias at each layer, this is a constant

> $n$ is the number of connections

> $Y$ is the output of the current neuron

> $\sum$ stands for sum

The equation you just read is called a weighed sum. We will take this weighted sum at each and every neuron as we pass information through the network. Then we will add what's called a bias to this sum. The bias allows us to shift the network up or down by a constant value. It is like the y-intercept of a line.

But that equation is the not complete one! We forgot a crucial part, **the activation function**. This is a function that we apply to the equation seen above to add complexity and dimensionality to our network. Our new equation with the addition of an activation function $F(x)$ is seen below.

> $Y =F((\sum_{i=0}^n w_i x_i) + b)$

Our network will start with predefined activation functions (they may be different at each layer) but random weights and biases. As we train the network by feeding it data it will learn the correct weights and biases and adjust the network accordingly using a technqiue called **backpropagation** (explained below). Once the correct weights and biases have been learned our network will hopefully be able to give us meaningful predictions. We get these predictions by observing the values at our final layer, the output layer.

# Building the neural network

We are going to build a neural network of three layers

Layer 1: This is our input layer and it will conist of 784 neurons. We use the flatten layer with an input shape
 of (28,28) to denote that our input should come in in that shape. The flatten means that our layer will reshape
  the shape (28,28) array into a vector of 784 neurons so that each pixel will be associated with one neuron.

Layer 2: This is our first and only hidden layer. The dense denotes that this layer will be fully connected and
each neuron from the previous layer connects to each neuron of this layer. It has 128 neurons and uses the rectify
 linear unit activation function.

Layer 3: This is our output later and is also a dense layer. It has 10 neurons that we will look at to determine
 our models output. Each neuron represnts the probabillity of a given image being one of the 10 different classes.
 The activation function softmax is used on this layer to calculate a probabillity distribution for each class.
 This means the value of any neuron in this layer will be between 0 and 1, where 1 represents a high probabillity
  of the image being that class.'''




```
'''Now it's time to build the model! We are going to use a keras sequential model with three different layers.
 This model represents a feed-forward neural network (one that passes values from left to right). We'll break
  down each layer and its architecture below.
'''
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),  # input layer (1)
    keras.layers.Dense(128, activation='relu'),  # hidden layer (2)
    # keras.layers.Dense(512, activation='relu'),  # hidden layer (2)
    keras.layers.Dense(10, activation='softmax') # output layer (3)
])
```

**Weight Reqularisation**

We can further reduce overfitting by introducing weight regularisation to our model. This is based on the assumption that simpler models are less likely to overfit than more complex models.  We define simpler models as models where the weights/ layer parameters have small weights. Largeer weights, more complex systems, are a sign the model is too specialised to the training data and it can provide varying ouputs when used in unseen data.


A "simple model" in this context is a model where the distribution of parameter values has less entropy (or a model with fewer parameters altogether, as we saw in the section above). Thus a common way to mitigate overfitting is to put constraints on the complexity of a network by forcing its weights to only take small values, which makes the distribution of weight values more "regular". This is called "weight regularization", and it is done by adding to the loss function of the network a cost associated with having large weights. This cost comes in two flavors:

L1 regularization, where the cost added is proportional to the absolute value of the weights coefficients (i.e. to what is called the "L1 norm" of the weights).
L2 regularization, where the cost added is proportional to the square of the value of the weights coefficients (i.e. to what is called the "L2 norm" of the weights). L2 regularization is also called weight decay in the context of neural networks. 


**Summary:  Weight regularasaiation punsishes the model for having large weight values. As models with smaller weights are less lileky to overfit.  This is achieved by adding a penalty to the loss function for having large weights. Large weights can cause the  model to be  too specialised too training data, meaning it will produce varting output on unseen data.**

```
model_diet = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),  # input layer (1)
    keras.layers.Dense(128, kernel_regularizer=keras.regularizers.l2(0.0005), activation='relu'),  # hidden layer (2)
#     # keras.layers.Dense(128, activation='relu'),  # hidden layer (2)
    keras.layers.Dense(10, kernel_regularizer=keras.regularizers.l2(0.0005), activation='softmax') # output layer (3)
])

# model_diet = tf.keras.models.Sequential()
# model_diet.add()


model_diet.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

print("Fit model on training data")

diet_history = model_diet.fit(train_images,
                            train_labels,
                            epochs=20,
                            validation_data=(val_images, val_labels),
                            )  # we pass the data, labels and epochs and watch the magic!

print(diet_history.history)

diet_accuracy = diet_history.history['accuracy']
diet_val_accuracy = diet_history.history['val_accuracy']
plot_accuracy(diet_accuracy, diet_val_accuracy, "diet Model")


# test_loss, test_acc = model_diet.evaluate(test_images,  test_labels, verbose=1)

# print('Test accuracy:', test_acc)
```

# Dropout


Droppout is another regularisation technique but works in a different way. During training time a set proportion(the drop out rate) of output features are set to 0, meaning less feature data is outputted from the layer. And during test time to make up for the removed features, The layeers output values are scaled down by the factor equal to the drop out rate. 

The core idea of this is to introudce
